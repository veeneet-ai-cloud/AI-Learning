```python
from regression.ages_net_worths import ageNetWorthData
import matplotlib.pyplot as plt
```

```python
ages_train, ages_test, net_worths_train, net_worths_test = ageNetWorthData()
```

```python
def studentRegression(ages_train, net_worths_train):
    from sklearn.linear_model import LinearRegression
    reg = LinearRegression()
    reg.fit(ages_train, net_worths_train)
    return reg
```

```python
reg = studentRegression(ages_train, net_worths_train)
```

```python
plt.clf()
plt.scatter(ages_train, net_worths_train, color="b", label="train data")
plt.scatter(ages_test, net_worths_test, color="r", label="test data")
plt.plot(ages_test, reg.predict(ages_test), color="black")
plt.legend(loc=2)
plt.xlabel("ages")
plt.ylabel("net worths")
plt.show()
```

```python
print ("Net worth prediction for 27 year old:", reg.predict([[27]]))
print ("slope:", reg.coef_)
print ("intercept:", reg.intercept_)
```

```python
print ("R-squared score:", reg.score(ages_test, net_worths_test))
print("R-squared score on training:", reg.score(ages_train, net_worths_train))
```
